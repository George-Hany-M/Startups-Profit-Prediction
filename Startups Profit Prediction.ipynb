# <font color='DarkSlateGrey'>‚öë Splitting the data and target</font>

```python
# X Data
X = data.drop(['Profit'], axis=1, inplace=False)
print('X Data is : \n' , X.head())
print()
print('X shape is : ' , X.shape)
print('-' * 100)
print()

# y Data
y = data['Profit']
print('y Data is : \n' , y.head())
print('y shape is : ' , y.shape)

#X = X.to_numpy()
#y = y.to_numpy()
#y = y.reshape(-1, 1)
```

# <font color='DarkSlateGrey'>‚öë Data Partitioning / Train Test Split</font>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle =True)

# Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

# <font color='DarkSlateGrey'>‚öë Data Scaling / Feature Scaling</font>

### <font color='SaddleBrown'>- <b>MinMaxScaler</b> is a preprocessing technique provided by Scikit-Learn that scales and transforms features in a dataset to a specified range, typically between 0 and 1.</font>
### <font color='SaddleBrown'>- This scaling is particularly useful for machine learning algorithms that require features to have similar ranges to prevent certain features from dominating the learning process.</font>
### <font color='SaddleBrown'>- <b>MinMaxScaler</b> can help improve the performance and convergence of models.</font>

```python
# MinMaxScaler for Data

scaler = MinMaxScaler()

# Fit the scaler on the training data
scaler.fit(X_train)

# Transform the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test) # Generalizing what you learned in "Training Data" ...

# It's important to fit the scaler only on the training data, not the test data.
```

# <font color='DarkSlateGrey'>‚öë Applying Linear Regression Model</font>

```python
Model = LinearRegression()
Model.fit(X_train_scaled, y_train)

# Calculating Details
print('Linear Regression Train Score is : ' , Model.score(X_train_scaled, y_train))
print('Linear Regression Test Score is : ' , Model.score(X_test_scaled, y_test))
print('Linear Regression Coef is : ' , Model.coef_)
print('Linear Regression intercept is : ' , Model.intercept_)

# Calculating Prediction
y_pred = Model.predict(X_test_scaled)
```

```python
y_predection = pd.DataFrame(data={'Predicted Profit ': y_pred.flatten()})
print(y_predection.head())
```

```python
print('Actual Profit')
y_test.head()
```

```python
Testing_Data_Model_score = Model.score(X_test_scaled, y_test)
print(f'Mode Score/Performance On Testing Data {Testing_Data_Model_score * 100:.2f}%')

Training_Data_Model_score = Model.score(X_train_scaled, y_train)
print(f'Mode Score/Performance On Training Data {Training_Data_Model_score * 100:.2f}%')
```

# <font color='DarkSlateGrey'>‚öë Plotting the performance of the model</font>

```python
# Model scores for each dataset
training_score = Training_Data_Model_score
testing_score = Testing_Data_Model_score

# Define datasets and corresponding scores
datasets = ['Training Data', 'Testing Data']
scores = [training_score, testing_score]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(datasets, scores, color=['blue', 'green'])
plt.ylim(0, 1)  # Set y-axis limits to be between 0 and 1 for accuracy percentage
plt.ylabel('Accuracy')
plt.title('Model Performance on Training and Testing Data')

# Adding values on top of the bars
for i, score in enumerate(scores):
    plt.text(i, score, f'{score * 100:.2f}%', ha='center', va='bottom')

plt.show()
```

# <font color='DarkSlateGrey'>‚öë Evaluation the model</font>

```python
# Calculating r2_score
r2_Score = r2_score(y_pred, y_test)
print('R2 Score Of Model is : ', r2_Score * 100)

# Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y_pred, y_test)
print('Mean Absolute Error Value is : ', MAEValue * 100)

# Calculating Mean Squared Error
MSEValue = mean_squared_error(y_pred, y_test)
print('Mean Squared Error Value is : ', MSEValue * 100)
```
